```python
import re

pattern = '^a...s$'
test_string = 'abyss'
result = re.match(pattern, test_string)

if result:
  print("Search successful.")
else:
  print("Search unsuccessful.")
```

```python
# Searching for the word "cat" in a sentence
text = "The cat is on the roof."
match = re.search(r'cat', text)

if match:
    print("Found:", match.group())  # Output: "Found: cat"
```

```python
import re

def contains_number(string):
    # Check if the string contains a digit
    if re.search(r'\d', string):
        return "The string contains a number."
    else:
        return "The string does not contain a number."

# Test the function
string = input("Enter a string: ")
print(contains_number(string))
```

```python
import re

def contains_uppercase(string):
    # Check if the string contains any uppercase letters (A-Z)
    if re.search(r'[A-Z]', string):
        return "The string contains an uppercase letter."
    else:
        return "The string does not contain any uppercase letters."

# Test the function
string = input("Enter a string: ")
print(contains_uppercase(string))
```

```python
import re

def contains_whitespace(string):
    # Check if the string contains any whitespace characters
    if re.search(r'\s', string):
        return "The string contains a whitespace character."
    else:
        return "The string does not contain any whitespace characters."

# Test the function
string = input("Enter a string: ")
print(contains_whitespace(string))
```

```python
import re

def find_all_a(string):
    # Find all occurrences of 'a'
    matches = re.findall(r'a', string)
    return f"Found {len(matches)} 'a' characters."

# Test the function
string = input("Enter a string: ")
print(find_all_a(string))
```
