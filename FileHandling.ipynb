# **File Handling**

So far we have seen different Python data types. We usually store our data in different file formats. In addition to handling files, we will also see different file formats(.txt, .json, .xml, .csv, .tsv, .excel) in this section. First, let us get familiar with handling files with common file format(.txt).

File handling is an import part of programming which allows us to create, read, update and delete files. In Python to handle data we use open() built-in function.

# Syntax
open('filename', mode) # mode(r, a, w, x, t,b)  could be to read, write, update

"r" - Read - Default value. Opens a file for reading, it returns an error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

"t" - Text - Default value. Text mode

"b" - Binary - Binary mode (e.g. images)

# **Opening Files for Reading**
The default mode of open is reading, so we do not have to specify 'r' or 'rt'.

```python
f = open('/content/demo.txt.txt')
print(f) # <_io.TextIOWrapper name='./files/reading_file_example.txt' mode='r' encoding='UTF-8'>
```

```python
txt = f.read()
print(type(txt))
print(txt)
f.close()
```

read(): read the whole text as string.

 If we want to limit the number of characters we want to read, we can limit it by passing int value to the read(number) method.

```python
f = open('/content/demo.txt.txt')
txt = f.read(15)
print(type(txt))
print(txt)
f.close()
```

readline(): read only the first line

```python
f = open('/content/demo.txt.txt')
line = f.readline()
print(type(line))
print(line)
f.close()
```

readlines(): read all the text line by line and returns a list of lines


```python
f = open('/content/demo.txt.txt')
lines = f.readlines()
print(type(lines))
print(lines)
f.close()
```

To write to an existing file, we must add a mode as parameter to the open() function:


"a" - append - will append to the end of the file, if the file does not it creates a new file.

Let us append some text to the file we have been reading:

```python
import pickle

# Python dictionary
data = {'id': 1, 'name': 'John Doe', 'age': 22, 'city': 'New York'}

# Serialize the dictionary
with open('data.pickle', 'wb') as file:
    pickle.dump(data, file)

print("Data serialized successfully.")
```

with open('data.pickle', 'wb') as file::


 This line opens a file named data.pickle in write-binary mode ('wb'). The with statement ensures that the file is properly closed after the block of code is executed.


'data.pickle' is the filename where the serialized data will be saved.
'wb' stands for write-binary mode. Since pickle writes in binary format, you need to open the file in binary mode.



pickle.dump(data, file): This function call serializes the data dictionary and writes it to the file.


data: The Python object to be serialized (in this case, a dictionary).
file: The file object where the serialized data is being written.


pickle.dump: Serializes the data object and stores it in the file specified by file.

```python
import pickle

# Deserialize the previously serialized dictionary
with open('data.pickle', 'rb') as file:
    loaded_data = pickle.load(file)

print("Data deserialized successfully.")
print(loaded_data)
```

```python
try:
    file = open("test.txt", "r")  # Open the file in read mode
    contents = file.read()        # Read the contents of the file
    file.close()                  # Close the file

    if "Python" in contents:      # Check if "Python" is in the contents
        print("The word 'Python' is found in the file.")
    else:
        print("The word 'Python' is not found in the file.")
except FileNotFoundError:          # Handle the case where the file is not found
    print("The file 'test.txt' could not be found.")
else:
    # If no exceptions occur, print a message
    print("The file 'test.txt' was successfully read.")

```

```python
def divide_numbers(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
    else:
        print(f"The result of {a} divided by {b} is: {result}")
    finally:
        print("This block always executes, regardless of exceptions.")
divide_numbers(10, 2)
divide_numbers(5, 0)
```

**Sometimes, it's useful to capture the exception as a string, either to log it, display it to the user, or for debugging purposes.**



**Converting Exceptions to Strings**

To convert an exception to a string, you can use the str() function or the traceback module for more detailed information.

1. **Basic Example Using str()**





```python
try:
    # Code that may raise an exception
    result = 10 / 0  # This will raise a ZeroDivisionError
except Exception as e:
    # Convert the exception to a string
    error_message = str(e)
    print("An error occurred:", error_message)
```

2. **Detailed Exception Information Using traceback**

```python
import traceback

try:
    # Code that may raise an exception
    result = 10 / 0  # This will raise a ZeroDivisionError
except Exception as e:
    # Get detailed exception information
    error_message = traceback.format_exc()
    print("Detailed error information:\n", error_message)
```

traceback.format_exc() is a function in Python's traceback module that provides a formatted string of the most recent exception traceback.

This is particularly useful for logging, debugging, or displaying detailed error information without stopping the program

# **Raising Exceptions**

Raising exceptions in Python using the raise keyword allows users to manually trigger exceptions based on certain conditions.

The raise keyword is useful for validating inputs, such as ensuring a divisor is not zero before performing division.

```python
# Taking input from the user
a = int(input("Enter value for a: "))
b = int(input("Enter value for b: "))

try:
    # Check if divisor is zero before performing division
    if b == 0:
        # Manually raise ZeroDivisionError
        raise ZeroDivisionError("Division by zero is not allowed.")

    # Perform division if no exception is raised
    c = a / b
    print("Result:", c)

except ZeroDivisionError as e:
    # Handle the raised ZeroDivisionError
    print("Exception occurred:", e)
```

```python
# Taking input for marks
marks = int(input("Enter marks: "))

try:
    # Validate if marks are within the range of 0 to 100
    if marks < 0 or marks > 100:
        # Manually raise an exception for invalid marks
        raise ValueError("Marks should be between 0 and 100.")

    # If marks are valid, print them
    print("Marks are:", marks)

except ValueError as e:
    # Handle the raised ValueError
    print("Exception occurred:", e)
```
